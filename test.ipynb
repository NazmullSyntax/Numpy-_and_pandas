```python
import numpy as np

data = pd.DataFrame({
    "Name": ["Alice", "Bob", np.nan, "Charlie"],
    "Age": [25, 30, 35,np.nan],
    "Salary": [50000, 60000, 70000, 80000]
})
```

```python
data
```

```python
data_dropped = data.dropna()
data_dropped
```
